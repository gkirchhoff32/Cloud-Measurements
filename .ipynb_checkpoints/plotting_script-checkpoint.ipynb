 ## Compare output from Deadtime Correction Technique (DCT) and Muller Correction

```python
import os
import sys

cwd = os.getcwd()
dirLib = cwd + r'/library'
if dirLib not in sys.path:
    sys.path.append(dirLib)
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
import pickle
from pathlib import Path

from load_ARSENL_data import set_binwidth
```

```python
home = str(Path.home())
pkl_dir = home + r'\OneDrive - UCB-O365\ARSENL\Experiments\Cloud Measurements\Sims\evaluation_loss\fit_figures'
muller_picklename = r'\muller_out_simnum5_downsamp5000_shots9.99E+02_usecom2_range975-1050.pkl'
dct_picklename = r'\DCT_out_dtimeTrue_simnum5_order18-23_shots9.99E+02_usecom2_range975-1050.pkl'
```

```python
c = 2.99792458e8  # [m/s] Speed of light
```

```python
# First load Muller correction and relevant variale

# Unpickle the data to DataFrame object
infile = open('{}/{}'.format(pkl_dir, muller_picklename), 'rb')
df_muller = pickle.load(infile)
infile.close()

infile = open('{}/{}'.format(pkl_dir, dct_picklename), 'rb')
df_dct = pickle.load(infile)
infile.close()
```

```python
dt = df_muller[0]
downsamp = df_muller[1]
t_min, t_max = df_muller[2], df_muller[3]
flight_time_LG = df_muller[4]
flight_time_HG = df_muller[5]
n_shots = df_muller[6]
deadtime = df_muller[7]
T_BS_LG, T_BS_HG = df_muller[8], df_muller[9]
photon_rate_arr_LG_muller = df_muller[10]
sim_num = df_muller[11]
t_fine_muller = df_muller[12]

discrete_loss = df_dct[0]
photon_rate_arr_LG_dct = df_dct[10]
t_fine_dct = df_dct[12]
fit_rate_seg = df_dct[13]
min_order = df_dct[14]
```

```python
ylim = np.array([975, 1050])  # [m]
```

```python
# Plotting histograms
res = dt * downsamp  # [s]
print('Processed Resolution: {:.3f} m ({:.3e} s)'.format(res * c / 2, res))
bin_array = set_binwidth(t_min, t_max+res, res)
n_LG, bins = np.histogram(flight_time_LG, bins=bin_array)
n_HG, __ = np.histogram(flight_time_HG, bins=bin_array)
binwidth = np.diff(bins)[0]
N_LG = n_LG / binwidth / n_shots  # [Hz] Scaling counts to arrival rate
N_HG = n_HG / binwidth / n_shots  # [Hz]
N_LG_muller = N_LG / (1 - N_LG*deadtime)  # [Hz] Muller correction applied directly to (unscaled) histogram
N_HG_muller = N_HG / (1 - N_HG*deadtime)  # [Hz]
N_LG_muller = N_LG_muller / T_BS_LG  # [Hz] Rescale histogram
N_HG_muller = N_HG_muller / T_BS_HG  # [Hz]
center_muller = 0.5 * (bins[:-1] + bins[1:])  # [s]
photon_rate_arr_muller = photon_rate_arr_LG_muller / T_BS_LG  # [Hz]
N_LG_muller = np.where(N_LG_muller==0, np.nan, N_LG_muller)
N_HG_muller = np.where(N_HG_muller==0, np.nan, N_HG_muller)

fig = plt.figure(figsize=(4, 8), dpi=400)
ax = fig.add_subplot(111)
ax.barh(center_muller * c / 2, N_LG_muller / 1e6, align='center', height=binwidth * c / 2, color='red', alpha=0.50, label='Muller Correction LG')
# ax.bar(center * c / 2, N_HG_muller / 1e6, align='center', width=binwidth * c / 2, color='black', alpha=0.50, label='Muller Correction (High Gain)')
ax.barh(center_muller * c / 2, N_LG / 1e6 / T_BS_LG, align='center', height=binwidth * c / 2, color='green', alpha=0.75, label='Low Gain (Scaled)')
ax.barh(center_muller * c / 2, N_HG / 1e6, align='center', height=binwidth * c / 2, color='blue', alpha=0.75, label='High Gain')
ax.barh(center_muller * c / 2, N_LG / 1e6, align='center', height=binwidth * c / 2, color='orange', alpha=0.75, label='Low Gain')
ax.plot(photon_rate_arr_muller / 1e6, t_fine_muller * c / 2, color='k', linestyle='--', label='Truth')
# ax.set_title('Arrival-Rate Fit Sim # {}'.format(sim_num))
ax.set_ylabel('Range [m]')
ax.set_xlabel('Photon Arrival Rate [MHz]')
plt.xscale('symlog')
plt.legend(prop={'size': 6})
plt.tight_layout()

fig = plt.figure(figsize=(4, 8), dpi=400)
ax = fig.add_subplot(111)
# ax.barh(center_muller * c / 2, N_HG_muller / 1e6, align='center', height=binwidth * c / 2, color='black', alpha=0.50, label='Muller Correction (High Gain)')
ax.plot(N_HG_muller / 1e6, center_muller * c / 2, color='m', linewidth=5, alpha=0.75, label='Muller Correction (HG)')
# ax.plot(N_LG_muller / 1e6, center_muller * c / 2, color='k', linewidth=5, alpha=0.75, label='Muller Correction (LG)')
# ax.barh(center_muller * c / 2, N_LG / 1e6 / T_BS_LG, align='center', height=binwidth * c / 2, color='green', alpha=0.75, label='Low Gain (Scaled)')
ax.barh(center_muller * c / 2, N_HG / 1e6, align='center', height=binwidth * c / 2, color='blue', alpha=0.75, label='HG Detections')
# ax.barh(center_muller * c / 2, N_LG / 1e6, align='center', height=binwidth * c / 2, color='orange', alpha=0.75, label='Low Gain')
ax.plot(photon_rate_arr_muller / 1e6, t_fine_muller * c / 2, color='k', linestyle='--', label='Truth')
# ax.set_title('Arrival-Rate Fit Sim # {}'.format(sim_num))
ax.set_ylabel('Range [m]')
ax.set_xlabel('Photon Arrival Rate [MHz]')
plt.xscale('symlog')
ax.set_xlim([-3e0, 5e3])
plt.legend(prop={'size': 6})
plt.tight_layout()

# Plotting error
LG_error_muller = np.abs(photon_rate_arr_muller - N_LG_muller)  # [Hz] Absolute error with Muller corrected histogram
HG_error_muller = np.abs(photon_rate_arr_muller - N_HG_muller)  # [Hz] Absolute error with Muller corrected histogram

# fig = plt.figure(figsize=(4, 8), dpi=400)
# ax = fig.add_subplot(111)
# ax.plot(LG_error_muller/1e6, center_muller * c / 2 / 1e3, color='green', linestyle='--', label='Muller LG error')
# ax.plot(HG_error_muller/1e6, center_muller * c / 2 / 1e3, color='blue', linestyle='--', label='Muller HG error')
# # ax.set_title('Arrival-Rate Fit Sim # {}'.format(sim_num))
# ax.set_ylabel('Range [km]')
# ax.set_xlabel('Absolute Error [MHz]')
# ax.tick_params(axis='y', which='minor')
# ax.semilogx()
# ax.set_xlim([1e-2, 1e3])
# ax.set_ylim(ylim / 1e3)
# plt.legend(prop={'size': 6})
# plt.tight_layout()
plt.show()
```

```python
# Plotting error
if discrete_loss:
    res = dt * downsamp  # [s]
else:
    res = dt  # [s]
print('Processed Resolution: {} m ({} s)'.format(res * c / 2, res))
bin_array = set_binwidth(t_min, t_max+res, res)
n_LG, bins = np.histogram(flight_time_LG, bins=bin_array)
binwidth = np.diff(bins)[0]
N_LG = n_LG / binwidth / n_shots  # [Hz] Scaling counts to arrival rate
photon_rate_arr_dct = photon_rate_arr_LG_dct / T_BS_LG
LG_error = np.abs(photon_rate_arr_dct - N_LG / T_BS_LG)  # [Hz] Absolute error with scaled LG histogram

fig = plt.figure(figsize=(4, 8), dpi=400)
ax = fig.add_subplot(111)
# ax.plot(LG_error/1e6, t_fine_dct * c / 2 / 1e3, color='green', linestyle='--', label='Scaled LG error')
ax.plot(np.abs(photon_rate_arr_dct - fit_rate_seg)/1e6, t_fine_dct * c / 2 / 1e3, color='r', linestyle='--', label='Fit error')
ax.set_title('Arrival-Rate Fit Sim # {}'.format(sim_num))
ax.set_xlabel('Range [km]')
ax.set_ylabel('Absolute Error [MHz]')
ax.tick_params(axis='y', which='minor')
ax.semilogx()
ax.set_xlim([1e-2, 1e3])
plt.legend(prop={'size': 6})
plt.tight_layout()

# Plotting histograms
use_bins = True
if use_bins:
    dsamp = 5000  # number of bins to downsample when plotting
    res_plot = dt * dsamp
else:
    res_plot = 2  # [m]
    res_plot = int(res_plot/c*2 // dt) * dt  # [s]
print('Figure Resolution: {} m ({} s)'.format(res_plot * c / 2, res_plot))
bin_array = set_binwidth(t_min, t_max+res_plot, res_plot)
n_LG, bins = np.histogram(flight_time_LG, bins=bin_array)
n_HG, __ = np.histogram(flight_time_HG, bins=bin_array)
binwidth = np.diff(bins)[0]
N_LG = n_LG / binwidth / n_shots  # [Hz] Scaling counts to arrival rate
N_HG = n_HG / binwidth / n_shots  # [Hz]
# try accomodating for combined high-gain and low-gain signals
photon_rate_arr_dct = photon_rate_arr_LG_dct / T_BS_LG
center_dct = 0.5 * (bins[:-1] + bins[1:])

fig = plt.figure(figsize=(4, 6), dpi=400)
ax = fig.add_subplot(111)
# ax.barh(center_dct * c / 2 / 1e3, N_LG / 1e6 / T_BS_LG, align='center', height=binwidth*c/2 / 1e3, color='green', alpha=0.75, label='Low gain counts (scaled)')
ax.barh(center_dct * c / 2 / 1e3, N_HG / 1e6, align='center', height=binwidth * c / 2 / 1e3, color='blue', alpha=0.75, label='High Gain')
ax.barh(center_dct * c / 2 / 1e3, N_LG / 1e6, align='center', height=binwidth * c / 2 / 1e3, color='orange', alpha=0.75, label='Low Gain')
ax.plot(photon_rate_arr_dct / 1e6, t_fine_dct * c / 2 / 1e3, color='k', linestyle='--', linewidth=7, label='Truth')
ax.plot(fit_rate_seg / 1e6, t_fine_dct * c / 2 / 1e3, color='c', linestyle='--', linewidth=7, alpha=0.75, label='DCT Fit')

# ax.set_title('Arrival-Rate Fit Sim # {}'.format(sim_num))
ax.set_xlabel('Range [km]')
ax.set_ylabel('Photon Arrival Rate [MHz]')
# props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
# ax.text(0.1, 0.90, 'Polynomial order: {}'.format(min_order), transform=ax.transAxes, fontsize=6, verticalalignment='top', bbox=props)
ax.set_xlim([1e0, 2e3])
ax.set_ylim([ylim[0]/1e3, 1020/1e3])
ax.semilogx()
plt.legend(prop={'size': 6})
plt.tight_layout()
plt.show()
```

```python
# Plot both Muller and DCT errors

fig = plt.figure(figsize=(8, 4), dpi=400)
ax = fig.add_subplot(111)
# ax.plot(t_fine_dct * c / 2, LG_error, color='orange', linestyle='--', label='Scaled LG error')
ax.plot(t_fine_dct * c / 2, np.abs(photon_rate_arr_dct - fit_rate_seg), color='r', linestyle='--', label='Fit error')
ax.plot(center_muller * c / 2, LG_error_muller, color='green', marker='.', linestyle='--', alpha=0.5, label='Muller LG error')
ax.plot(center_muller * c / 2, HG_error_muller, color='blue', marker='.', linestyle='--', alpha=0.5, label='Muller HG error')
ax.set_title('Arrival-Rate Fit Sim # {}'.format(sim_num))
ax.set_xlabel('Range [m]')
ax.set_ylabel('Absolute Error [Hz]')
ax.tick_params(axis='y', which='minor')
ax.semilogy()
ax.set_ylim([1e4, 1e11])
ax.set_xlim(ylim)
plt.legend(prop={'size': 6})
plt.tight_layout()
```

 ## Plotting Histograms

```python
import numpy as np
import time
import pickle
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path

from load_ARSENL_data import load_INPHAMIS_data, set_binwidth, data_dir, fname, picklename
```

```python
start = time.time()
# Constants
c = 299792458  # [m/s] Speed of light

# Parameters
limit_shots = False

# window_bnd = [32e-9, 38e-9]  # [s] Set temporal boundaries for binning
window_bnd = np.array([900, 1200])  # [m] Set boundaries for binning to exclude outliers
window_bnd = window_bnd / c * 2  # [s] Convert from range to tof
dt = 25e-12  # [s] Resolution
deadtime = 29.1e-9  # [s] Deadtime interval (25ns for sim, 29.1ns for SPCM)
if limit_shots:
    use_shots = 250

t_min = window_bnd[0]  # [s]
t_max = window_bnd[1]  # [s]

home = str(Path.home())
data_dir = home + r'\OneDrive - UCB-O365\ARSENL\Experiments\Cloud Measurements\Sims\saved_sims'
fname_hg = r'\\simnum_5_nshot1.00E+03_useHGTrue_T0.95.nc'
fname_lg = r'\\simnum_5_nshot1.00E+03_useHGFalse_T0.05.nc'
```

```python
ds_hg = xr.open_dataset(data_dir + fname_hg)
ds_lg = xr.open_dataset(data_dir + fname_lg)

cnts_hg = ds_hg.time_tag
cnts_lg = ds_lg.time_tag
T_BS_HG = ds_hg.T_BS.values
T_BS_LG = ds_lg.T_BS.values
if limit_shots:
    lim_shot_idx_hg = (abs(ds_hg.time_tag_sync_index - use_shots)).argmin().values.item()
    lim_shot_idx_lg = (abs(ds_lg.time_tag_sync_index - use_shots)).argmin().values.item()
    cnts_hg = cnts_hg[:lim_shot_idx_hg]
    cnts_lg = cnts_lg[:lim_shot_idx_lg]

flight_time_hg = cnts_hg * dt  # [s]
flight_time_lg = cnts_lg * dt  # [s]
# Exclude specified t.o.f. bins
flight_time_hg = flight_time_hg[np.where((flight_time_hg >= window_bnd[0]) & (flight_time_hg < window_bnd[1]))]
flight_time_lg = flight_time_lg[np.where((flight_time_lg >= window_bnd[0]) & (flight_time_lg < window_bnd[1]))]
if limit_shots:
    n_shots_hg = len(ds_hg.sync_index[:use_shots])
    n_shots_lg = len(ds_lg.sync_index[:use_shots])
else:
    n_shots_hg = len(ds_hg.sync_index)
    n_shots_lg = len(ds_lg.sync_index)

### Histogram of time of flight ###
fig = plt.figure(figsize=(4,6), dpi=400)
ax1 = fig.add_subplot(111)
bin_avg = 500
res = dt * bin_avg
bin_array = set_binwidth(t_min, t_max, res)
n_hg, bins = np.histogram(flight_time_hg, bins=bin_array)
n_lg, bins = np.histogram(flight_time_lg, bins=bin_array)
print('Histogram plot time elapsed: {:.3} sec'.format(time.time() - start))
binwidth = np.diff(bins)[0]
N_hg = n_hg / binwidth / n_shots
N_lg = n_lg / binwidth / n_shots
print('Number of shots: {}'.format(n_shots))
center = 0.5 * (bins[:-1]+bins[1:])
ax1.barh(center*c/2/1e3, N_hg/1e6, align='center', height=binwidth*c/2/1e3, color='b', alpha=0.5, label='High Gain (unscaled)')
ax1.barh(center*c/2/1e3, N_lg/1e6, align='center', height=binwidth*c/2/1e3, color='orange', alpha=0.5, label='Low Gain (unscaled)')
ax1.set_ylabel('Range [km]')
ax1.set_xlabel('Arrival rate [MHz]')
ax1.set_xscale('log')
ax1.set_ylim(window_bnd*c/2/1e3)
plt.tight_layout()
plt.legend()

fig = plt.figure(figsize=(4,6), dpi=400)
ax1 = fig.add_subplot(111)
ax1.barh(center*c/2/1e3, N_lg/1e6/T_BS_LG, align='center', height=binwidth*c/2/1e3, color='orange', alpha=1, label='Low Gain (rescaled)')
# ax1.barh(center*c/2/1e3, N_hg/1e6/T_BS_HG, align='center', height=binwidth*c/2/1e3, color='b', alpha=0.5, label='High Gain (rescaled)')
ax1.set_ylabel('Range [km]')
ax1.set_xlabel('Arrival rate [MHz]')
ax1.set_xscale('log')
ax1.set_ylim(np.array([950, 1050])/1e3)
ax1.set_xlim(3e-1, 5e2)
plt.tight_layout()
plt.legend()
plt.show()
plt.show()

fig = plt.figure(figsize=(4,4), dpi=400)
ax1 = fig.add_subplot(111)
# ax1.barh(center*c/2/1e3, N_lg/1e6/T_BS_LG, align='center', height=binwidth*c/2/1e3, color='orange', alpha=0.5, label='Low Gain (rescaled)')
ax1.barh(center*c/2/1e3, N_hg/1e6/T_BS_HG, align='center', height=binwidth*c/2/1e3, color='b', alpha=0.7, label='High Gain (rescaled)')
ax1.set_ylabel('Range [km]')
ax1.set_xlabel('Arrival rate [MHz]')
ax1.set_xscale('log')
ax1.set_ylim(np.array([950, 1050])/1e3)
ax1.set_xlim(3e-1, 5e2)
plt.tight_layout()
plt.legend()
plt.show()
plt.show()

```
