```python
import numpy as np
import matplotlib.pyplot as plt
import imageio as iio
from pathlib import Path

home = str(Path.home())
```

```python
open_dir = home + r'\OneDrive - UCB-O365\ARSENL\Experiments\Cloud Measurements\Scope Alignment\03062024'

# open tower image of 800mm FL Newtonian
img_twr_800 = iio.v3.imread(open_dir + r'\tower_edge_3_G260_E1111_NEWT.jpg')

# open tower image of Cassegrain
img_twr_cass = iio.v3.imread(open_dir + r'\tower_edge_1_G260_E250_CASS.jpg')

fig = plt.figure(dpi=400)
ax1 = fig.add_subplot(211)
ax1.imshow(img_twr_800, cmap='gray')
ax2 = fig.add_subplot(212)
ax2.imshow(img_twr_cass, cmap='gray')
# ax.set_ylim([400, 200])
plt.show()
```

```python
xsec_800 = img_twr_800[500, :]
xpix = np.arange(0, len(img_twr_800[0, :]))
xloc = xpix * 4.7  # [um] transforming pixels to dimensions

xsec_cass = img_twr_cass[500, :]


fig = plt.figure(dpi=400)
ax1 = fig.add_subplot(221)
ax1.plot(xloc, xsec_800, '.')
ax1.set_xlim([min(xloc), max(xloc)])
ax1.set_xlabel('Dimension [um]')
ax1.set_ylabel('Value [8-bit]')
ax1.set_title('Newtonian (800mm FL)')

ax2 = fig.add_subplot(223)
ax2.imshow(img_twr_800, cmap='gray')
ax2.set_ylim([600, 400])
ax2.set_xlabel('Pixels')
ax2.set_ylabel('Pixels')
ax2.axhline(500, alpha=0.35)

ax3 = fig.add_subplot(222)
ax3.plot(xloc, xsec_cass, '.')
ax3.set_xlim([min(xloc), max(xloc)])
ax3.set_xlabel('Dimension [um]')
ax3.set_title('Cassegrain')
# ax3.set_ylabel('Value [8-bit]')

ax4 = fig.add_subplot(224)
ax4.imshow(img_twr_cass, cmap='gray')
ax4.set_ylim([600, 400])
ax4.set_xlabel('Pixels')
# ax4.set_ylabel('Pixels')
ax4.axhline(500, alpha=0.35)

plt.tight_layout()
plt.show()
```

```python
# Look at individual lights (not true point source since they aren't far enough away)

open_dir = home + r'\OneDrive - UCB-O365\ARSENL\Experiments\Cloud Measurements\Scope Alignment'

# cassegrain image of tower light
img_spot_cass = iio.v3.imread(open_dir + r'\03052024\spot_from_2.jpg')

# newtonian image of tower light
img_spot_800 = iio.v3.imread(open_dir + r'\03062024\tower_light_1_G260_E1111_NEWT.jpg')

fig = plt.figure(dpi=400)
ax1 = fig.add_subplot(121)
ax1.imshow(img_spot_800, cmap='gray')
ax1.set_title('Newtonian (800mm FL)')
ax2 = fig.add_subplot(122)
ax2.imshow(img_spot_cass, cmap='gray')
ax2.set_title('Cassegrain')
plt.tight_layout()
plt.show()
```

```python
F_newt = 800e-3  # [m]
F_cass = 4  # [m]
s = 3690  # [m] Distance of tower
h = 0.4  # [m] diameter of lamp

M_newt = F_newt / (s-F_newt)
M_cass = F_cass / (s-F_cass)

hp_newt = M_newt * h  # [m]
hp_cass = M_cass * h  # [m]

print('Newtonian: {:.3e} m'.format(hp_newt))
print('Cassegrain: {:.3e} m'.format(hp_cass))
```
